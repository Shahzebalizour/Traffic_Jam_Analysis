## 📋 Column Names:

Certainly! Here's an explanation of each column in simple English with emojis:

1. **Hour (⏰):** This column tells us the time when the observation was recorded. It's like a clock, showing what time of the day it is (e.g., 7:00 AM, 8:30 AM).

2. **Immobilized Bus (🚌):** This column counts how many buses were unable to move or got stuck on the road. It helps us understand if buses are facing issues.

3. **Broken Truck (🚚):** It keeps track of how many trucks have broken down on the road. It helps monitor incidents involving trucks.

4. **Vehicle Excess (🚗):** This column counts instances when there are too many vehicles on the road. It helps us see if traffic congestion is due to an excessive number of vehicles.

5. **Accident Victim (🚑):** It records the number of people who have been victims of accidents on the road. It helps track the safety of the roads.

6. **Running Over (🏃):** This column counts incidents where someone has been run over on the road. It's a measure of pedestrian safety.

7. **Fire Vehicles (🚒):** It tracks the number of times fire vehicles like fire trucks were involved in incidents. It's important for emergency response monitoring.

8. **Freight Incidents (📦):** Counts the occurrences related to freight transportation, like shipments. It helps us see if there are issues with freight on the road.

9. **Dangerous Freight Incidents (☠️):** Similar to the previous column but focuses on incidents involving dangerous or hazardous freight, which could be harmful.

10. **Lack of Electricity (💡):** Records instances where there was a lack of electricity, which can impact traffic signals and lighting on the road.

11. **Fire (🔥):** Counts the number of fire-related incidents on the road, such as vehicle fires or other types of fires.

12. **Point of Flooding (💧):** It tracks locations where flooding has occurred on the road due to heavy rain or other factors.

13. **Manifestations (📣):** This column counts manifestations or public demonstrations on the road. It helps monitor public events that might affect traffic.

14. **Trolleybus Network Defect (🚎):** Records incidents related to issues in the trolleybus network, a type of public transportation system.

15. **Tree on the Road (🌳):** Counts instances where trees have fallen onto the road, potentially blocking traffic.

16. **Semaphore Off (🚦):** It tracks the times when traffic signals (semaphores) were turned off, which can affect traffic flow.

17. **Intermittent Semaphore (🔴🟢):** Records incidents when traffic signals were intermittent, showing both red and green lights. It's a signal issue.

18. **Slowness in Traffic (%) (🚦📉):** This is the target column. It represents the percentage of traffic slowness on the road. It tells us how slow the traffic is at a given time, expressed as a percentage. For example, if it's 10%, it means traffic is moving at 10% of the normal speed.
